```python
import os

# 디렉토리 경로 설정
directory_path = r'C:\source\kamp_ex\train'

path_arr = []
# 디렉토리 항목 나열
with os.scandir(directory_path) as it:
    for entry in it:
        # 항목의 이름 출력
        print(f'Name: {entry.name}')
        
        # 항목의 전체 경로 출력
        print(f'Path: {entry.path}')
        
        # 항목이 파일인지 여부 출력
        if entry.is_file():
            print('This is a file.')
        elif entry.is_dir():
            print('This is a directory.')
            with os.scandir(entry) as ent:
                for i in ent:
                    print(i)
                    path_arr.append(i)
        
        # 항목의 상태 출력
        print(f'Stats: {entry.stat()}')
        
        print('---')
```

```python
import os

def list_all_files_with_paths(directory):
    file_paths = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            full_path = os.path.join(root, file)
            file_paths.append(full_path)
    return file_paths

# 사용 예제
directory = r'C:\source\kamp_ex\train\ok'
files_with_paths = list_all_files_with_paths(directory)
print(files_with_paths)
```
