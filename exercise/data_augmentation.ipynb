- 이미지 데이터 증강 기법 data augmentation 오그맨테이션
- 이미지 데이터 증강은 컴퓨터 비전 분야에서 가장 많이 사용됩니다. 이미지에 다양한 변형을 적용하여 새로운 이미지를 생성합니다.

1. 회전(Rotation): 이미지를 일정 각도로 회전합니다.
1. 이동(Translation): 이미지를 수평 또는 수직으로 이동시킵니다.
1. 확대/축소(Scaling): 이미지를 크기를 키우거나 줄입니다.
1. 뒤집기(Flipping): 이미지를 수평 또는 수직으로 뒤집습니다.
1. 왜곡(Shearing): 이미지를 기울여 변형합니다.
1. 자르기(Cropping): 이미지의 일부분을 자릅니다.
1. 색상 변형(Color Jittering): 이미지의 색상, 밝기, 대비, 채도를 변경합니다.
1. 노이즈 추가(Adding Noise): 이미지에 랜덤한 노이즈를 추가합니다.
1. 컷아웃(Cutout): 이미지의 일부분을 검은색 또는 다른 색으로 덮습니다.
