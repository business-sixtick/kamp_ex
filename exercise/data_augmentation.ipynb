```python

image = s.cv.imread('../image/KEMP_IMG_DATA_1.png')
height, width = image.shape[:2]
print(height, width)

for i in range(10):
    angle = s.random() * 10 - 5
    rotation_matrix = s.cv.getRotationMatrix2D((width / 2, height / 2), angle, 1)
    rotation_matrix[0][2] += s.random() * 72 - 36 #   
    rotation_matrix[1][2] += s.random() * 48 - 24 #   
    # print(rotation_matrix, type(rotation_matrix), rotation_matrix[0][2])
    rotated_image = s.cv.warpAffine(image, rotation_matrix, (width, height))
    # cv2.imshow('Rotated Image', rotated_image)
    gray_image = s.cv.cvtColor(rotated_image, s.cv.COLOR_BGR2GRAY)
    s.image_center_show('Rotated Image', gray_image)
    s.cv.waitKey(1000)
s.cv.destroyAllWindows()
```
