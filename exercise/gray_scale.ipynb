```python
import cv2
import matplotlib.pyplot as plt
```

```python


#  
image = cv2.imread('../image/KEMP_IMG_DATA_1.png')

#  
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print(gray_image.shape)
plt.imshow(gray_image, cmap='gray')
#   
# cv2.imshow('Grayscale Image', gray_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
img_ok = cv2.imread('../image/KEMP_IMG_DATA_1.png', cv2.IMREAD_GRAYSCALE)
img_bad = cv2.imread('../image/KEMP_IMG_DATA_Error_2.png', cv2.IMREAD_GRAYSCALE)
img_bad2 = cv2.imread('../image/KEMP_IMG_DATA_Error_12.png', cv2.IMREAD_GRAYSCALE)

fig, axs = plt.subplots(1, 3)
fig.set_figheight(6)
fig.set_figwidth(12)
axs[0].set_xlabel('ok', fontsize=32, color='red')
axs[1].set_xlabel('bad', fontsize=32, color='red')
axs[2].set_xlabel('bad', fontsize=32, color='red')
axs[0].imshow(img_ok, cmap='gray')
axs[1].imshow(img_bad, cmap='gray')
axs[2].imshow(img_bad2, cmap='gray')

print(img_ok.shape)
```
